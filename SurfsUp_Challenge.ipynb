```python
%matplotlib inline
from matplotlib import style
import matplotlib.pyplot as plt

# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import numpy as np
import pandas as pd
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
results_june = session.query(Measurement).filter(extract("month", Measurement.date) == 6)
print(results_june.all())

# first_row = session.query(Measurement).first()
# first_row.__dict__
```

```python
#  3. Convert the June temperatures to a list.
temps_june = []
for temp in results_june:
    temps_june.append(temp.tobs)
print(temps_june)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
df_june = pd.DataFrame(temps_june, columns=['june_temp'])
df_june.head()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
df_june.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
results_december = session.query(Measurement).filter(extract("month", Measurement.date) == 12)
print(results_december.all())
```

```python
# 7. Convert the December temperatures to a list.
temps_december = []
for temp in results_december:
    temps_december.append(temp.tobs)
print(temps_december)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
df_december = pd.DataFrame(temps_december, columns=['december_temp'])
df_december.head()
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
df_december.describe()
```

```python
x_labels = ["June Temperature"]
fig, ax = plt.subplots()
ax.boxplot(temps_june, labels=x_labels)
# Add the title, y-axis label and grid.
ax.set_ylabel('Temperature(Fahrenheit)')
ax.set_yticks(np.arange(60, 90, step=2.0))
ax.grid()
plt.savefig("./June_Boxplot.png")
plt.show()
```

```python
x_labels = ["December Temperature"]
fig, ax = plt.subplots()
ax.boxplot(temps_december, labels=x_labels)
# Add the title, y-axis label and grid.
ax.set_ylabel('Temperature(Fahrenheit)')
ax.set_yticks(np.arange(54, 84, step=2.0))
ax.grid()
plt.savefig("./December_Boxplot.png")
plt.show()
```

```python
# Create a box-and-whisker plot for the urban fare data.
x_labels = ["June", "December"]
fare_count_data = [temps_june, temps_december]
fig, ax = plt.subplots(figsize=(8,6))
ax.boxplot(fare_count_data, labels=x_labels)

# Add the title, y-axis label and grid.
ax.set_title('June and December Temperature Data')
ax.set_ylabel('Temperature(Fahrenheit)')
ax.set_yticks(np.arange(54, 86, step=2.0))
ax.grid()
plt.savefig("./June_December_Boxplots.png")
plt.show()
```
